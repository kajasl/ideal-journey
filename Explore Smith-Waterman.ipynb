```python
def count_length(tracks):
    mask = numpy.core.defchararray.count(tracks.polyline.values.astype(str), '[') -1
    

    point_list = pd.DataFrame(tracks, columns=["polyline", "value"])
    
    point_list["val"] = mask
    #for i in range(len(point_list)):
        #val = len(point_list.loc[i]["polyline"])
     #   point_list.at[i, "polyline"] = val
    return point_list
```

```python
def create_list_of_coordinates (dataset):
    polylines = pd.DataFrame()
    lon, lat, id = [], [], []

    for i in range(len(dataset)):
        try:
            lst = json.loads(dataset["polyline"].iloc[i])
            index, latte, longi = zip(*lst)
            lon = lon+ list(longi)
            lat=lat+list(latte)
            id = id+[dataset.index[i] for j in range(len(lst))]
        except:
            print(dataset.index[i])

    polylines["lat"]=lat
    polylines["lon"]=lon
    polylines["id"]=id
    return polylines

poly = create_list_of_coordinates(res)
#print(poly)
```
