```python
import pandas as pd
import json
import numpy
import geopy.distance
import plotly.express as px
```

```python
dataset = pd.read_csv("Needleman_Wunsch_result_1.csv")
dataset.sort_values(by=['value'], ascending=False, inplace=True)
res = dataset.loc[dataset['value'] > 15]
print(len(res))
```

```python
def create_list_of_coordinates (dataset):
    polylines = pd.DataFrame()
    lon, lat, id = [], [], []

    for i in range(len(dataset)):
        try:
            lst = json.loads(dataset["polyline"].iloc[i])
            index, longi, latte = zip(*lst)
            lon = lon+ list(longi)
            lat=lat+list(latte)
            id = id+[dataset.index[i] for j in range(len(lst))]
        except:
            print(dataset.index[i])

    polylines["lat"]=lat
    polylines["lon"]=lon
    polylines["id"]=id
    return polylines

poly = create_list_of_coordinates(res)
print(poly[:5])
```

```python
fig = px.line_mapbox(poly, lat="lat", lon="lon", color="id", zoom=3, height=700)

fig.update_layout(mapbox_style="stamen-terrain", mapbox_zoom=12, mapbox_center_lat = 41.14,
    margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
