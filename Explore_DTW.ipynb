```python
def create_list_of_coordinates (dataset):
    polylines = pd.DataFrame()
    lon, lat, id = [], [], []

    for i in range(len(dataset)):
        try:
            lst = json.loads(dataset["polyline"].iloc[i])
            index, latte, longi = zip(*lst)
            lon = lon+ list(longi)
            lat=lat+list(latte)
            id = id+[dataset.index[i] for j in range(len(lst))]
        except:
            print(dataset.index[i])

    polylines["lat"]=lat
    polylines["lon"]=lon
    polylines["id"]=id
    return polylines
```
